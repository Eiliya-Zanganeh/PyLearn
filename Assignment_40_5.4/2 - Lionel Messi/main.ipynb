```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Lionel Messi Goals.csv')
```

```python
df.head()
```

```python
df['Date'] = df['Date'].apply(lambda x: x[-4:])
df['Date'] = df['Date'].astype(int)
df.head()
```

```python
result = df['Date'].value_counts().reset_index()
result = result.sort_values(by=['Date'], ascending=False)
result.head()
```

```python
plt.bar(result['Date'], result['count'])
plt.xticks(rotation=90)
plt.ylabel('Goals Count')
plt.show()
```

```python
result = df[df['Goal Type'] == 'Penalty']
result = result['Goal Method'].value_counts().reset_index()
result.head()
```

```python
plt.pie(result['count'], labels=result['Goal Method'], autopct='%1.1f%%')
plt.title('Goal Methods in Penalty')
plt.show()
```

```python
df['Minute'] = df['Minute'].apply(lambda x: eval(x))
df['Minute'] = df['Minute'].astype(int)
df.head(10)
```

```python
result = df[df['Minute'] >= 90]
result = result['Final Game Result'].value_counts().reset_index()
result.head()
```

```python
plt.pie(result['count'], labels=result['Final Game Result'], autopct='%1.1f%%')
plt.title('Result in overtime')
plt.show()
```
