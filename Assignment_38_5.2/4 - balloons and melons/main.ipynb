```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
balloons = pd.DataFrame(columns=['width', 'height', 'weight'])

balloons['width'] = np.random.normal(10, 1, 500)
balloons['height'] = np.random.normal(20, 1, 500)
balloons['weight'] = np.random.normal(2, 1, 500)
```

```python
melons = pd.DataFrame(columns=['width', 'height', 'weight'])

melons['width'] = np.random.normal(20, 1, 500)
melons['height'] = np.random.normal(10, 1, 500)
melons['weight'] = np.random.normal(5, 1, 500)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(balloons['width'], balloons['height'], balloons['weight'], color='red', marker='o', label='Balloons')

ax.scatter(melons['width'], melons['height'], melons['weight'], color='yellow', marker='^', label='Melons')

ax.set_xlabel('Width')
ax.set_ylabel('Height')
ax.set_zlabel('Weight')

ax.legend()

plt.show()
```
