```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('COVID-19-time-series-clean-complete.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year_Month'] = df['Date'].map(lambda date: 100 * date.year + date.month)

df.head()
```

```python
df = df[df['Year_Month'] == 202004]

df.head()
```

```python
result = df.groupby('Country/Region')['New cases'].sum().reset_index()

result.head()
```

```python
result = result.sort_values('New cases', ascending=False)[:8]
result.head()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(result['Country/Region'], result['New cases'])
plt.title('8 countries with highest corona patients')
plt.xlabel('Country/Region')
plt.ylabel('Patients count')
plt.show()
```
