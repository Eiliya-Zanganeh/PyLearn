```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Billionaire.csv')
```

```python
df.head()
```

```python
df = df.dropna()
df.isnull().sum()

df['NetWorth'] = df['NetWorth'].str.strip('$')
df['NetWorth'] = df['NetWorth'].str.strip('B')
df['NetWorth'] = df['NetWorth'].astype(float)

df.head()
```

```python
ages = df['Age'].astype(int).tolist()
print(f'Var: {np.var(ages)}')
print(f'Mean: {np.mean(ages)}')
```

```python
result = df.groupby('Country')['NetWorth'].sum().reset_index()

result = result.sort_values(by='NetWorth', ascending=False)
result.head()
```

```python
plt.figure(figsize=(18, 3))
plt.bar(result['Country'], result['NetWorth'])
plt.xticks(rotation=90)
plt.ylabel('Billionaires')
plt.show()
```

```python
result = df[df['Country'] == 'United States'].copy()
result = result.sort_values(by=['NetWorth'], ascending=False)[:10]
result.head(10)
```

```python
plt.bar(result['Name'], result['NetWorth'])
plt.xticks(rotation=90)
plt.ylabel('Billionaires')
plt.show()
```

```python
result = df.groupby('Country')['NetWorth'].sum().reset_index()

result = result.sort_values(by='NetWorth', ascending=True)[:10]
result.head()
```

```python
plt.bar(result['Country'], result['NetWorth'])
plt.xticks(rotation=90)
plt.ylabel('Billionaires')
plt.show()
```
