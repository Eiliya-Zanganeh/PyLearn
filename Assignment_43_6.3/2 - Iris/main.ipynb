```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

from k_nearest_neighbors import KNearestNeighbors
```

```python
dataset = load_iris()
X = dataset.data
Y = dataset.target
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=42)
```

```python
model = KNearestNeighbors(x_train, y_train)
```

```python
predict = model.predict(x_test, k=3)
cm = confusion_matrix(y_test, predict)
accuracy = accuracy_score(y_test, predict)
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title(f'Confusion Matrix: Accuracy: {round(accuracy * 100)}%, K: 3')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
predict = model.predict(x_test, k=5)
cm = confusion_matrix(y_test, predict)
accuracy = accuracy_score(y_test, predict)
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title(f'Confusion Matrix: Accuracy: {round(accuracy * 100)}%, K: 5')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
