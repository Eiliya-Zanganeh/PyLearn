```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_dataset(n=100):
    X = np.random.random(n) * 5
    Y = X * 2 + np.random.random(n)
    return X, Y


X, Y = generate_dataset()
plt.scatter(X, Y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=42)
```

```python
x_train = x_train.reshape(-1, 1)
w = np.matmul(x_train.T, x_train)
w = np.linalg.inv(w)
w = np.matmul(w, x_train.T)
w = np.matmul(w, y_train)
w
```

```python
plt.scatter(X, Y, color='green')
plt.plot(x_train, x_train * w, color='blue')
plt.plot(x_train, x_train * 1, color='red')
plt.plot(x_train, x_train * 3.5, color='red')
plt.plot(x_train, x_train * 5, color='red')
plt.show()
```

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x_train.reshape(-1), y_train)
plt.scatter(X, Y, color='orange')
plt.plot(x_train, x_train * w, color='blue')
plt.plot(x_train, x_train * slope, color='green')
plt.show()
```
