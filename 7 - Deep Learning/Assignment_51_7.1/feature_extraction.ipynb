```python
import os

path = "Datasets/PersianFace/"
folders = [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]
print(folders)
```

```python
import pandas as pd
from deepface import DeepFace
from tqdm import tqdm
import cv2 as cv

cols = ['label', 'file_name']
cols.extend(["feature_{}".format(i) for i in range(512)])
df = pd.DataFrame(columns=cols)

for person in tqdm(folders):
  folder = path + person
  for file in os.listdir(folder):
      file_path = os.path.join(folder, file)
      if os.path.isfile(file_path):
          data = [person, file_path]
          img = cv.imread(file_path)
          features = DeepFace.represent(img_path=img, model_name="ArcFace", enforce_detection=False)
          data.extend(features[0]['embedding'])
          df.loc[len(df)] = data
```
