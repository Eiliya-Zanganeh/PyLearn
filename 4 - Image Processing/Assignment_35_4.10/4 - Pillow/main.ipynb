```python
from PIL import Image, ImageFont, ImageDraw, ImageOps
import arabic_reshaper
from bidi.algorithm import get_display
import matplotlib.pyplot as plt
```

```python
original_image = Image.open('inputs/img.png')
```

```python
image = original_image.copy()
draw = ImageDraw.Draw(image)
font = ImageFont.truetype("./Far_Nazanin.ttf", 20)

text = "من کیم؟"

reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text)

text_position = (165, 120)
text_color = (0, 0, 0)
draw.text(text_position, bidi_text, font=font, fill=text_color)

plt.imshow(image)
plt.show()
```

```python
r, g, b = original_image.split()

r_hist = r.histogram()
g_hist = g.histogram()
b_hist = b.histogram()

plt.plot(r_hist, color="red")
plt.plot(g_hist, color="green")
plt.plot(b_hist, color="blue")
plt.show()
```

```python
r_eq = ImageOps.equalize(r)
g_eq = ImageOps.equalize(g)
b_eq = ImageOps.equalize(b)

equalized_color_image = Image.merge("RGB", (r_eq, g_eq, b_eq))

plt.imshow(equalized_color_image)
plt.show()
```

```python
r, g, b = equalized_color_image.split()

r_hist = r.histogram()
g_hist = g.histogram()
b_hist = b.histogram()

plt.plot(r_hist, color="red")
plt.plot(g_hist, color="green")
plt.plot(b_hist, color="blue")
plt.show()
```

```python
gray = original_image.convert("L")

plt.imshow(gray, cmap="gray")
plt.show()
```

```python
histogram = gray.histogram()

plt.plot(histogram, color="gray")

plt.show()
```

```python
equalized_image = ImageOps.equalize(gray)

plt.imshow(equalized_image, cmap="gray")
plt.show()
```
